```python
# importing beautiful soup
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/Template:COVID-19_pandemic_data'
page = requests.get(url)
print(page.content)
```

```python
# parsing the html
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Finding the wikitable sortable class and printing the content
right_table=soup.find('table', class_='wikitable plainrowheaders sortable')
print(right_table)
```

```python
#Create lists for as many columns in the table
L1 = []
L2 = []
L3 = []
L4 = []
```

```python
# Initiate the count to 0 to find the a tag and extracting the name of country with get_text
count=0
for row in right_table.findAll('tr'):
  count=count+1
  if count>=3:
    L1.append(row.findAll('a')[0].get_text())
    cells=row.findAll('td')
  if len(cells)==4:
    L2.append(cells[0].find(text=True).rstrip("\n"))
    L3.append(cells[1].find(text=True).rstrip("\n"))
    L4.append(cells[2].find(text=True).rstrip("\n"))
    
```

```python
#Printing the length of all the lists
#print(len(L1))
#print(len(L2))
#print(len(L3))
#print(len(L4))
```

```python
L1=L1[0:239]
```

```python
import pandas as pd
df=pd.DataFrame()
df['Location']   = L1
df['Cases']    = L2
df['Deaths']     = L3
df['Recoveries']  = L4

df

```

```python
# Exporting the file into CSV
df.to_csv("COVID cases by countries.csv")
```

```python
df['Cases']=df['Cases'].astype(float)
df['Deaths']=df['Deaths'].astype(float)
df['Recoveries']=df['Recoveries'].astype(float)
```
